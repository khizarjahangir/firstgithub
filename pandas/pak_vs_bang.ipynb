## Pak vs Bangladesh graph (together)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame

df3 = DataFrame({'ban_y': ban_y, 'Bangladesh': ban_v, 'Pakistan': pak_v})

# method 1
# df3.plot()
# plt.title("Number of malnurished people")
# plt.xlabel("Years")
# plt.ylabel("Value in millions")

# method 2
ax = df3.plot()
ax.set_xticks(df3.index)
ax.set_xticklabels(df3.ban_y, rotation=90)
ax.set_xlabel("Years")
ax.set_ylabel("Value in millions")
ax.set_title("Number of malnurished people")
```

```python
# df3.plot.bar()
# plt.title("Number of malnurished people")
# plt.xlabel("Years")
# plt.ylabel("Value in millions")

# method 2
ax = df3.plot.bar(figsize=(8,6),legend=True, fontsize=11)
ax.set_xticks(df3.index)
ax.set_xticklabels(df3.ban_y, rotation=90)
ax.set_xlabel("Years")
ax.set_ylabel("Value in millions")
ax.set_title("Number of malnurished people")

# save as png
plt.savefig("figure.png") 
```
