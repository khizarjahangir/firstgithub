```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#canvas
sns.set(style='whitegrid')
# import data
df1 = pd.read_csv("Chilla_data2_for_plots.csv")

#defining a color palette
# p= sns.color_palette('rocket_r')

#line plot
sns.relplot(data=df1,x="Age (years)-Float/Int", y="Marital Status?", 
            hue="Are you Vaccinated?",size="Gender",
           kind="line",size_order=["T1","T2"],
           height=5, aspect=.75, facet_kws=dict(sharex=False))
```

```python
sns.boxplot(data=df1,x="Age (years)-Float/Int", y="Marital Status?", 
            hue="Are you Vaccinated?",
            palette=["m", "g"])
sns.despine(offset=10, trim=True)
```
